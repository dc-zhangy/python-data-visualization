在```seaborn```中，有多种不同的方式来展示包含了分类数据的变量关系。正如```relplot()```和```scatterplot()/lineplot()```之间的关系一样，我们可以使用```catplot()```函数来描述分类数据，也可以使用更多坐标轴级别的绘图函数来完成这些任务。```catplot()```提供了对这些axes-level的函数的整合，将他们放在了一个更高级别的统一的接口之中。
- 分类散点图
  - ```stripplot()```: 或```catplot(kind="strip")```
  - ```swarmplot()```: 或```catplot(kind="swarm")```
-  分类分布图
  - ```boxplot()```: 或```catplot(kind="box")```
  - ```violinplot()```: 或```catplot(kind="violin")```
  - ```boxenplot(): ```或```catplot(kind="boxen")```
-  分类统计估计图
   - ```pointplot()```: 或```catplot(kind="point")```
   - ```barplot()```: 或```catplot(kind="bar")```
   - ```countplot()```: 或```catplot(kind="count")```

````python
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(style="ticks", color_codes=True)
````

````python
tips = sns.load_dataset("tips")
tips.head()
````

# 分类散点图
## 散点图
`catplot(kind=“strip”)`默认

````python
sns.catplot(x="day", y="total_bill", data=tips,height=4,aspect=1.5)
````

`jitter`参数控制着偏移量的大小，或者我们可以直接禁止他们偏移

````python
sns.catplot(x="day", y="total_bill",jitter=False, data=tips,height=4,aspect=1.5)
````

````python
## 蜂群图
`catplot(kind=“swarm”)`
<p>避免了散点之间的重合。它提供了更好的方式来呈现观测点的分布,
<p>但是它仅适用于较小的数据集
````

````python
sns.catplot(x="day", y="total_bill", kind="swarm", data=tips,height=4,aspect=1.5);
````

- `hue`参数:增加一个新的维度
- 但是分类图不支持`size`和`style`.

````python

sns.catplot(x="day", y="total_bill", hue="sex", kind="swarm", data=tips,height=4,aspect=1.5)
````

- 关于x（分类型数据）轴的分类值的顺序，函数会自己推断。
- 如果数据是pandas的Categorical类型，那么默认的分类顺序可以在pandas中设置；
- 如果数据看起来是数值型的（比如1，2，3…），那他们也会被排序。
- 即使我们使用数字作为不同分类的标签，它们仍然是被当做分类型数据按顺序绘制在分类变量对应的坐标轴上的：



````python
# 注意，2和4之间的距离与1和2之间的距离是一样的，它们是不同的分类，只会排序，但是并不会改变它们在坐标轴上的距离
sns.catplot(x="size", y="total_bill", kind="swarm", data=tips.query("size != 3"),height=4,aspect=1.5);
````

`order`参数:指定分类

````python
sns.catplot(x="smoker", y="tip", order=["No", "Yes"], data=tips,height=4,aspect=1.5)
````

有些时候把分类变量放在垂直坐标轴上会更有帮助（尤其是当分类名称较长或者分类较多时）,只需要交换x和y分配的变量即可

````python
sns.catplot(x="total_bill", y="day", hue="time", kind="swarm", data=tips,height=4,aspect=1.5)
````

# 分类分布图
当数据集的大小越来越大，分类散点图在表现不同分类的观测值的分布信息时就越发显得捉襟见肘。此时，有其他的方式：

## 箱线图
`catplot(kind=“box”)`

展现了：离群点、上界、上四分位数、均值、中位数、下四分位数、下界、离群点


````python
sns.catplot(x="day", y="total_bill", kind="box", data=tips,height=4,aspect=1.5)
````

- `hue`参数:增加一个新的维度

````python
sns.catplot(x="day", y="total_bill", kind="box",hue="smoker", data=tips,height=4,aspect=1.5)
````

上面的图中，默认了hue参数对应的变量smoker与坐标轴上的分类变量day是相互嵌套的（如，周四吸烟、周四不吸烟），这种操作叫**“dodging”**。

**如果不是这种情形，应该禁用dodging，看下面的例子：**

**hue的Weekend和day不是相互嵌套的**

````python
tips["Weekend"]= tips["day"].isin(["Sat","Sun"])
tips.head()
````

**× 错误的图**

````python
sns.catplot(kind="box", x="day", y="total_bill", hue="Weekend", data=tips,height=4,aspect=1.5)
````

**√ 正确的图（禁用dodge，dodge=False）**

````python
sns.catplot(kind="box", x="day", y="total_bill", hue="Weekend",dodge=False, data=tips,height=4,aspect=1.5)
````

**box的进阶：**<p>
` catplot(kind=“boxen”)`<p>
与箱线图相似但是能展示更多关于数据分布形状的信息，它对大数据更加友好：

````python
sns.catplot(kind="boxen", x="day", y="total_bill", data=tips,height=4,aspect=1.5)
````

## 小提琴图
`catplot(kind=“violin”)`<p>
另一种方法是violinplot()，它将箱线图与核密度估计过程结合了起来：

````python
sns.catplot(kind="violin", y="day", x="total_bill",hue="time", data=tips)
````

参数`split=Ture`<p>
当一个额外的分类变量仅有2个水平时，我们也可以将它赋给hue参数，并且设置split=True，这样我们可以更加充分地利用空间来表达更多信息：

````python
sns.catplot(kind="violin", y="day", x="total_bill",hue="time",split=True, data=tips)
````

# 分类统计估计图
在某些应用场景中，相对于展示每类的分布情况，更像展示每类的数据的集中趋势估计（统计量，如均值、中位数、方差等）。


````python
titanic = sns.load_dataset("titanic")
titanic.head()
````

## 条形图：
`catplot(kind=“bar”)`<p>
在`seaborn`中，`barplot()`函数在整个数据集上运行，并且应用一个函数来获得那些统计量（默认为均值）。当每个分类中有多个观测值时，它还可以通过自助采样法计算出一个置信区间，并且通过误差棒的方式绘制出来

````python
sns.catplot(kind="bar", x="sex", y="survived", hue="class", data=titanic,height=4,aspect=1.5)
````

`catplot(kind=“count”)`：展示每个分类下观测值的数量

````python
sns.catplot(x="deck", kind="count",color='c',edgecolor="0.6",palette="ch:.25", data=titanic);
````

# 绘制“宽表”数据
宽表类数据是指二维的数据，比如`pandas`的`DataFrame`或者`numpy`的二维数组。这些对象可以直接传递给data参数。

````python
iris = sns.load_dataset("iris")
iris.head()
````

````python
sns.catplot(data=iris, orient="h", kind="box")
````

接受类似向量的数据输入（比如`pandas`的`Series`、`numpy`的一维数组）

````python
sns.violinplot(x=iris.species, y=iris.sepal_length)
````

使用更底层的matplotlib设置项控制图形的大小和形状

````python
f, ax = plt.subplots(figsize=(7, 3))
sns.countplot(y="deck", data=titanic, color="c");
````

# 展示多重关系

````python
sns.catplot(x="day", y="total_bill", hue="smoker",
            col="time", aspect=.6,
            kind="swarm", data=tips);
````

使用FacetGrid对象定制更多细节

````python
g = sns.catplot(x="fare", y="survived", row="class",
                kind="box", orient="h", height=1.5, aspect=4,
                data=titanic.query("fare > 0"))
g.set(xscale="log")
````
