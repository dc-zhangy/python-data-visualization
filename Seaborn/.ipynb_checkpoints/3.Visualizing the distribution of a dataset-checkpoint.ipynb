```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy import stats

sns.set(color_codes=True)
```

# 绘制单变量分布
`distplot()`

```python
x = np.random.normal(size=1000)
sns.distplot(x)
```

## 直方图
地毯图`rugplot()`会在每个观测值的位置添加一个垂直的线段小标记

```python
# 高度和下方标记的密集度是成正比的
sns.distplot(x, kde=False, rug=True,bins=20)
```

## 核密度估计

```python
sns.distplot(x, hist=False, rug=True)
```

 对于直方图，KDE图的绘制有更高的计算复杂度。在KDE的计算中，首先每个观测值会被替换为一条以该观测点为中心的正态曲线：

```python
x = np.random.normal(0, 1, size=30)
bandwidth = 1.06 * x.std() * x.size ** (-1 / 5.)
support = np.linspace(-4, 4, 200)

kernels = []
for x_i in x:
    kernel = stats.norm(x_i, bandwidth).pdf(support)
    kernels.append(kernel)
    plt.plot(support, kernel, color='r')

sns.rugplot(x, color=".2", linewidth=3);
```

接下来，将这些曲线求和以计算支撑网格中每个点的密度值。然后将结果曲线归一化，以使其下方的面积等于1

```python
from scipy.integrate import trapz
density = np.sum(kernels, axis=0)
density /= trapz(density, support)
plt.plot(support, density)
```

如果我们直接用`kdeplot()`函数，那我们会得到同样的结果。`distplot()`中集成了这个函数，但是当我们只想绘制密度估计图时，`kdeplot()`能提供更直接的接口以及更多的选项

```python
sns.kdeplot(x, shade=True)
```

`kdeplot()`中的`bw`参数控制着估计值与真实数据之间的贴近程度，就像直方图中的`bin`一样。它与我们的KDE图的宽度相关。它提供了默认的规则来确定一个取值，不过有些时候我们自己尝试一些更大或更小的值也许会很有帮助

```python
sns.kdeplot(x, label="bw: default")
sns.kdeplot(x, bw=.2, label="bw: 0.2")
sns.kdeplot(x, bw=2, label="bw: 2")
plt.legend();
```

高斯（正态）KDE过程的性质扩展了原数据集中的最大值和最小值。我们可以通过`cut`参数控制曲线超过极值的距离，然而这只会改变曲线的绘制，而不会改变它拟合、计算的过程：

```python
sns.kdeplot(x, shade=True, cut=0)
sns.rugplot(x)
```

拟合指定的分布

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma);
```

# 绘制二元分布
`jointplot()`

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 2000)
df = pd.DataFrame(data, columns=["x", "y"])
df.head()
```

## 散点图

```python
sns.jointplot(x="x", y="y", data=df)
```

## 回归估计

```python
sns.jointplot(x="x", y="y", data=df,kind='reg')
```

## 六边图
`jointplot(kind="hex")`<p>
六边图（Hexbin），可以视为二维世界中的直方图，因为它展示了每个小六边形中观测点的数量。这种图在大数据集上表现最佳。我们使用`matplotlib`的p`lt.hexbin`函数以及`seaborn`的`jointplot(kind="hex")`都可以绘制它

```python
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k")
```

## 核密度估计
`jointplot(kind="kde")`

```python
sns.jointplot(x="x", y="y", data=df, kind="kde");
```

我们也可以直接使用`kdeplot()`来绘制二维的核密度估计图。这允许我们直接在某个指定的`matplotlib`坐标轴上（包括已存在的坐标轴体系）绘制这类图形。作为对比，在`jointplot()`中我们只能创建一个新的图来生成核密度估计图。

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```

```python
f, ax = plt.subplots(figsize=(6, 6))
sns.kdeplot(df.x, df.y, ax=ax)
sns.rugplot(df.x, color="g", ax=ax)
sns.rugplot(df.y, vertical=True, ax=ax);
```
