```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
sns.set(color_codes=True)
tips = sns.load_dataset("tips")
tips['big_tip']=tips['smoker']=='No'
tips.head()
```

# 绘制线性回归模型
`regplot()`<p>
`lmplot()`<p>
 在最基本的调用过程中，他们都会画出关于x、y两个变量的散点图，同时用数据拟合一个y ~ x的模型出来，并将对应的直线和95%的置信区间绘制出来   

```python
sns.regplot(x="total_bill", y="tip", data=tips);
```

```python
sns.lmplot(x="total_bill", y="tip", data=tips);
```

我们注意到除了图片的形状略有差异，其他地方都是一致的。需要知道的是，他们之间最主要的区别在于`regplot()`中的x, y参数接受多种数据类型，包括`numpy`数组、`pandas`序列（`Series`），或者将`pandas DataFrame`传递给`data`参数。作为对比，`lmplot()`的`data`参数是不能为空的，同时x和y参数必须以字符串形式指定。这种数据格式（这里是指`regplot()`支持而`lmplot()`不支持的类似一维数组的数据格式）被称为“long-form data”或“tidy data”。除了这一输入格式的灵活性以外，`regplot()`仅提供了`lmplot()`特性的一部分，所以我们用后者来演示它们的使用。

seaborn支持其中一个变量属于离散变量的情况，不过这种数据集产生的散点图效果往往一般：

```python
sns.lmplot(x="size", y="tip", data=tips);
```

增加一些随机的偏移量会让这些分布看起来更清晰，需要注意的是这些偏移量仅仅会影响散点图的效果，不会对拟合的回归线产生干扰：

```python
sns.lmplot(x="size", y="tip", data=tips, x_jitter=.05);
```

另一个选择是将每个离散桶内的观测值隐藏起来，用代表集中趋势的统计量以及置信区间作为替代：

```python
sns.lmplot(x="size", y="tip", data=tips, x_estimator=np.mean,height=4,aspect=1.5);
```

# 拟合不同的模型

```python
anscombe = sns.load_dataset("anscombe")
anscombe.head()
```

## 多项式回归
`order`

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, ci=None, scatter_kws={"s": 80},height=4,aspect=1.5);
```

## 稳健回归
`robust=True`

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           robust=True, ci=None, scatter_kws={"s": 80},height=4,aspect=1.5)
```

## 逻辑回归
`logistic=True`

```python
sns.lmplot(x="total_bill", y="big_tip", data=tips,
           logistic=True, y_jitter=.03,height=4,aspect=1.5);
```

## LOWESS

```python
sns.lmplot(x="total_bill", y="tip", data=tips, lowess=True,height=4,aspect=1.5);
```

## 残差检验
`residplot()`<p>
函数可以检查一个简单的回归模型对于某个数据集是否合适。它先拟合一个简单线性回归模型并移除它，然后将每个观测点与预测值的残差画出来。理想情况下，这些残差应该随机地分布在x轴上下方：

```python
sns.residplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
              scatter_kws={"s": 80});
```

# 考虑其他变量

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,height=4,aspect=1.5)
```

通过标记样式来区分不同的水平

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips,
           markers=["o", "x"], palette="Set1",height=4,aspect=1.5)
```

通过更多的子图（行、列）来引入新的变量

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", col="time", data=tips,height=4,aspect=1.3);
```
