### DataFrame.plot

```python
np.random.seed(123)

df = pd.DataFrame({'A': np.random.randn(365).cumsum(0),
                   'B': np.random.randn(365).cumsum(0) + 20,
                   'C': np.random.randn(365).cumsum(0) - 20},
                 index = pd.date_range('1/1/2017', periods=365))
df.head()
```

```python
df.plot()
```

```python
df.plot();  # Notice the semicolon it suppresses unwanted output
```

```python
df.plot('A', 'B', kind = 'scatter');
```

We can also choose the plot kind by using the `DataFrame.plot.kind` methods instead of providing the `kind` keyword argument.

`kind` :
- `'line'` : line plot (default)
- `'bar'` : vertical bar plot
- `'barh'` : horizontal bar plot
- `'hist'` : histogram
- `'box'` : boxplot
- `'kde'` : Kernel Density Estimation plot
- `'density'` : same as 'kde'
- `'area'` : area plot
- `'pie'` : pie plot
- `'scatter'` : scatter plot
- `'hexbin'` : hexbin plot

```python
df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
```

```python
ax = df.plot.scatter('A', 'C', c='B', s=df['B'], colormap='viridis')
ax.set_aspect('equal')
```

```python
df.plot.box();
```

```python
df.plot.hist(alpha=0.7);
```

```python
df.plot.kde();
```

# Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib notebook
```

```python
np.random.seed(1234)

v1 = pd.Series(np.random.normal(0, 10, 10000), name='v1')
v2 = pd.Series(2*v1 + np.random.normal(60, 15, 10000), name='v2')
```

```python
# plot a kernel density estimation over a stacked barchart
plt.figure()
plt.hist(v1, alpha=0.7, bins=np.arange(-50,150,5), label='v1');
plt.hist(v2, alpha=0.7, bins=np.arange(-50,150,5), label='v2');
plt.legend();
```

```python
plt.figure()
plt.hist([v1, v2], histtype='barstacked', density=True);
v3 = np.concatenate((v1, v2))
sns.kdeplot(v3);
```

```python
plt.figure()
sns.distplot(v3, hist_kws={'color': 'Teal'}, kde_kws={'color': 'Navy'});
```

```python
sns.jointplot(v1, v2, alpha=0.4);
```

```python
grid = sns.jointplot(v1, v2, alpha=0.4);
grid.ax_joint.set_aspect('equal')
```

```python
sns.jointplot(v1, v2, kind='hex');
```

```python
# set the seaborn style for all the following plots
sns.set_style('white')

sns.jointplot(v1, v2, kind='kde', space=0);
```
