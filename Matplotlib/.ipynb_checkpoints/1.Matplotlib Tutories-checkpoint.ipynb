# 折线图

```python
#%matplotlib notebook
%matplotlib inline
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
mpl.rcParams['font.sans-serif']=['SimHei'] #用来正常显示中文标签
mpl.rcParams['axes.unicode_minus']=False #用来正常显示负号
```

## 默认参数作图

```python
X = np.linspace(-np.pi, np.pi, 20, endpoint=True)#生成-π到+π的20个元素的等差数列
plt.figure()#创建一张图
C,S = np.cos(X), np.sin(X)#生成x的正弦余弦函数并赋值给C，S
plt.plot(X,C)#画出x与C的图像
plt.plot(X,S)#画出x与S的图像
plt.savefig("pi.jpg",dpi=150,bbox_inches='tight')#保存图片，分辨率为150,让图片没有多余的空白
#plt.show()#展示图像 jupyter中可以不用这一命令
```

## 修改常规默认设置

```python
X = np.linspace(-np.pi, np.pi, 20, endpoint=True)#生成-π到+π的20个元素的等差数列
C,S = np.cos(X), np.sin(X)#生成x的正弦余弦函数并赋值给C，S
plt.figure(figsize=(8,4), dpi=80)#设置图表的宽高比为10:6，设置dpi为80

#设置标题
plt.title('正弦曲线')
plt.xlabel('X axis')
plt.ylabel('Y axis')

#改变颜色、标记、线宽、线条样式
plt.plot(X, C, marker='o',color="blue", linewidth=1, linestyle="--",label="$cos(t)$")#设置余弦函数颜色为蓝色，线宽2.5，样式为虚线,label为图例，
plt.plot(X, S, marker='*',color="red",  linewidth=2.5, linestyle="-",label="$sin(t)$")#设置余弦函数颜色为蓝色，线宽2.5，样式为实线

#改变坐标轴刻度范围
plt.xlim(X.min()*1.5, X.max()*1.5)#x坐标轴最小值、最大值分别设置为X最小值和最大值的1.1
plt.ylim(C.min()*1.5, C.max()*1.5)#y坐标轴最小值、最大值分别设置为X最小值和最大值的1.1

#改变坐标轴显示
plt.xticks([-np.pi, -np.pi/2, 0, np.pi/2, np.pi], [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'$+\pi$'])#设置将刻度映射为具体标签
plt.yticks([-1, 0, +1],[r'$-1$', r'$0$', r'$+1$'])#显示+-1,0的刻度

#设置图例
plt.legend(loc='best', frameon=False)#loc和frameon指定图例位置和图例边框有无
```

## 修改其他设置
### 移动坐标轴

```python
X = np.linspace(-np.pi, np.pi, 20, endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.figure(figsize=(8,4), dpi=80)
plt.plot(X, C, marker='o',color="blue", linewidth=1, linestyle="--",label="$cos(t)$")
plt.plot(X, S, marker='*',color="red",  linewidth=2.5, linestyle="-",label="$sin(t)$")

ax = plt.gca()                                            # get current axis 获得坐标轴对象

ax.spines['right'].set_color('none') 
ax.spines['top'].set_color('none')         # 将右边 上边的两条边颜色设置为空 其实就相当于抹掉这两条边

ax.xaxis.set_ticks_position('bottom')   
ax.yaxis.set_ticks_position('left')          # 指定下边的边作为 x 轴   指定左边的边为 y 轴

ax.spines['bottom'].set_position(('data', 0))   #指定 data  设置的bottom(也就是指定的x轴)绑定到y轴的0这个点上
ax.spines['left'].set_position(('data', 0))

```

### 任意位置添加文本

```python
X = np.linspace(-np.pi, np.pi, 20, endpoint=True)
C,S = np.cos(X), np.sin(X)
plt.figure(figsize=(8,4), dpi=80)
plt.plot(X, C, marker='o',color="blue", linewidth=1, linestyle="--",label="$cos(t)$")
plt.plot(X, S, marker='*',color="red",  linewidth=2.5, linestyle="-",label="$sin(t)$")
plt.text(-2.,0.5,"$y=cos(t)$")
plt.text(0,0,"$y=sin(t)$")
```

### 子图、对数坐标轴

```python
y = np.random.normal(loc=0.5, scale=0.4, size=1000)
y = y[(y > 0) & (y < 1)]
y.sort()
x = np.arange(len(y))

# plot with various axes scales
plt.figure(figsize=(20,4))

# linear
plt.subplot(1,3,1)#子图
plt.plot(x, y)
plt.yscale('linear')
plt.title('linear')

# log
plt.subplot(1,3,2)
plt.plot(x, y)
plt.yscale('log')
plt.title('log')
plt.grid(True)#网格线

# symmetric log
plt.subplot(1,3,3)
plt.plot(x, y - y.mean())
plt.yscale('symlog', linthreshy=0.05)
plt.title('symlog')
plt.grid(True)
```

### 函数间颜色填充

```python
plt.figure(figsize=(20,4))

#填充函数与坐标轴
plt.subplot(1,3,1)
x = np.linspace(0, 5 * np.pi, 1000)
y1 = np.sin(x)
y2 = np.sin(2 * x)
plt.plot(x, y1,x, y2)
plt.fill(x, y1, color = "g",alpha=0.3)
plt.fill(x, y2, color = "b",alpha=0.3)

#填充函数与函数之间
plt.subplot(1,3,2)
plt.plot(x, y1,x, y2)
plt.fill_between(x, y1, y2, facecolor = "red",alpha=0.3)

#条件填充
plt.subplot(1,3,3)
plt.plot(x, y1,x, y2)
plt.fill_between(x, y1, y2, where= y1 >= y2, facecolor = "blue",alpha=0.3)
plt.fill_between(x, y1, y2, where= y2 > y1, facecolor = "yellow",alpha=0.3)
```

# 散点图

```python
N = 50
x = np.random.rand(N)
y = np.random.rand(N)

area = np.pi * (15 * np.random.rand(N))**2 # 0 to 15 point radiuses
color = 2 * np.pi * np.random.rand(N)

#默认参数
plt.figure(figsize=(20,8))
plt.subplot(2,3,1)
plt.scatter(x, y)

#修改标记、颜色、大小
plt.subplot(2,3,2)
plt.scatter(x, y,marker='*',color='black',s=50)
plt.title('散点图')
plt.xlabel('X axis')
plt.ylabel('Y axis')

#修改标记颜色、大小（用列表形式）
plt.subplot(2,3,3)
plt.scatter(x, y, s=area,c=color ,alpha=0.5)# alpha颜色通道从 0 (透明) to 1 (不透明)

plt.subplot(2,3,4)
plt.scatter(x, y, s=area, alpha=0.5,edgecolors='black',facecolors="None")
#图例
plt.subplot(2,3,5)
z = np.random.rand(N)
plt.scatter(x,y,marker='o',label='y')
plt.scatter(x,z,marker='x',label='z')
plt.legend(loc='best')
```

# 柱状图

```python
x=np.arange(1,7)
y=x
name=['北京','上海','天津','重庆','广州','南京']
#默认参数
plt.figure(figsize=(20,8),dpi=300)
plt.subplot(2,3,1)
plt.subplots_adjust(wspace =0.2, hspace =0.3)#调整子图间距

plt.bar(x,y)
plt.title('默认参数')

#修改宽度，颜色，坐标轴显示
plt.subplot(2,3,2)
plt.bar(x,y,width=0.5,edgecolor='black',facecolor='None',align='center')
plt.xticks(x,name)
plt.xlabel('城市')
plt.ylabel('数量')
plt.title('修改宽度、颜色、坐标轴显示')

# 堆积柱形图
plt.subplot(2,3,3)
z=[1,6,3,5,1,7]
plt.bar(x, y, width = 0.3, color='b',label='男性')
plt.bar(x, z, width = 0.3, bottom=y, color='r',label='女性')
plt.bar(x, z[::-1], width = 0.3, bottom=y+z, color='yellow',label='中性')
plt.xticks(x,name)
plt.legend()
plt.title('堆积柱形图')

#横向堆积图
plt.subplot(2,3,4)
plt.barh(x, y, color='b',label='男性')
plt.barh(x, z, left=y, color='r',label='女性')
plt.barh(x, z[::-1], left=y+z, color='yellow',label='中性')
plt.yticks(x,name)
plt.legend()
plt.title('横向堆积图')

#并列柱形图
plt.subplot(2,3,5)
width=0.3
plt.bar(x, y, width = width, color='b',label='男性')
plt.bar(x+width, z, width = width, color='r',label='女性')
plt.xticks(x+0.5*width,name)
plt.legend()
plt.title('并列柱形图')

#显示数据标签
plt.subplot(2,3,6)
plt.bar(x,y,width=0.5,edgecolor='black',facecolor='None',align='center')
plt.xticks(x,name)
plt.xlabel('城市')
plt.ylabel('数量')
for i in range(len(x)):
    plt.text(np.squeeze(x[i]),np.squeeze(y[i])+0.1,str(np.squeeze(y[i])))
plt.ylim(0,7)
plt.title('显示数据标签')


plt.savefig("bar.jpg",dpi=150,bbox_inches='tight')
```

# 子图subplot三种布局
## 均等分割布局

```python
plt.figure('Figure Object 1',facecolor = 'lightgray')

for i in range(2):
    for j in range(3):
        k = i * 3 + j + 1
        plt.subplot(2,3,k)
        plt.xticks(())
        plt.yticks(())
        plt.text(0.5, 0.5, str(k), ha='center', va='center',
                 size = 36, alpha=0.5)      
plt.show()
```

## 自定义布局

```python
import matplotlib.gridspec as gridspec
plt.figure()
gs = gridspec.GridSpec(3, 3,hspace=0.5,wspace=0.5)#生成一个3*3的图,调整子图间距

Y = np.random.normal(loc=0.0, scale=1.0, size=10000)
X = np.random.random(size=10000)

plt.subplot(gs[0, 1:])#画上边的图
plt.hist(X, bins=100)

plt.subplot(gs[1:, 0])#画左边的图
plt.hist(Y, bins=100, orientation='horizontal')

plt.subplot(gs[1:, 1:])#画底下的图
plt.scatter(X, Y)
```

# 箱线图

```python
normal = np.random.normal(loc=0.0, scale=1.0, size=10000)
random = np.random.random(size=10000)
gamma = np.random.gamma(2, size=10000)

#画一个
plt.figure(figsize=(16,4))
plt.subplot(1,3,1)
plt.boxplot(random )

#画多个
plt.subplot(1,3,2)
plt.boxplot([random,normal,gamma])#圆圈表示异常值
#不显示异常值
plt.subplot(1,3,3)
plt.boxplot([random,normal,gamma],showfliers=False)#圆圈表示异常值
```
